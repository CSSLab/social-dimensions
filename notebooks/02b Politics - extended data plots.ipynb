```python
import matplotlib.gridspec as gridspec

def render_left_right_polarization_appendix(partisan_dimen):

    figsize = (7,6)
    fig = plt.figure(figsize=figsize)
    gs0 = gridspec.GridSpec(2,2, figure=fig)
    gs0.update(wspace=0.4, hspace=0.3)

    gs_a = gs0[0, 0]
    gs_b = gs0[0, 1]
    
    gs_c = gs0[1, 0]
    gs_d = gs0[1, 1]
    
    ax_line = pol.render_abs_z_line_plot(fig, gs_a, partisan_dimen)
    plots.add_subplot_label(ax_line, "a", x=-40)
    ax_line2 = pol.render_abs_z_line_plot(fig, gs_b, partisan_dimen, metric="num_comments")
    plots.add_subplot_label(ax_line2, "b",x=-40)
    
    ax_line.tick_params(axis='x', which='major', rotation=45)
    ax_line2.tick_params(axis='x', which='major', rotation=45)
    ax_line.set_xlabel("")
    ax_line2.set_xlabel("")
    ax_line.margins(x=0)
    ax_line2.margins(x=0)

    s = 4.5
#     ax_cohort_l = pol.render_cohort_plot(fig, gs_f, partisan_dimen, "left", point_size=s)
#     plots.add_subplot_label(ax_cohort_l, "a")
#     ax_cohort_r = pol.render_cohort_plot(fig, gs_g, partisan_dimen, "right", sharey=ax_cohort_l, point_size=s)
#     plots.add_subplot_label(ax_cohort_r, "b")
    
    ax_pol_l = pol.render_explained_polarization(fig, gs_c, partisan_dimen, "left",fontsize=10)
    plots.add_subplot_label(ax_pol_l, "c", x=-40)
    ax_pol_r = pol.render_explained_polarization(fig, gs_d, partisan_dimen, "right", sharey=ax_pol_l, fontsize=10)
    plots.add_subplot_label(ax_pol_r, "d", x=-40)
    
#     ax_cohort_r.get_legend().remove()
    ax_pol_r.get_legend().remove()
    
    # keep ticks
#     ax_cohort_r.yaxis.set_tick_params(labelleft=True)
    ax_pol_l.yaxis.set_tick_params(labelleft=True)
    ax_pol_r.yaxis.set_tick_params(labelleft=True)
    #ax_pol_l.text(-2.6, 0.4, "Change in polarization", va='center', rotation='vertical')
    
    l_col = [0.2298057 , 0.29871797, 0.75368315, 1.        ]
    r_col = [0.70567316, 0.01555616, 0.15023281, 1.        ]
    for l, r in [(ax_pol_l, ax_pol_r)]: # , 
        l.annotate("Left", (1, 1), xytext=(-10, -10), textcoords='offset points',
                   xycoords='axes fraction', ha='right', va='top', color=l_col,
                   weight='bold')
        r.annotate("Right", (0, 1), xytext=(10, -10), textcoords='offset points',
                   xycoords='axes fraction', ha='left', va='top', color=r_col,
                   weight='bold')

        
    return fig, []

render_left_right_polarization_appendix("partisan")
```

```python

def render_original_political_activity(partisan_dimen):

    figsize = (10,9)
    fig = plt.figure(figsize=figsize)
    
    gs0 = gridspec.GridSpec(1, 1, figure=fig)

    gs_a = gs0[:, :]
    
    pol.render_partisan_activity_plot(fig, gs_a, partisan_dimen, just_first_part=False)

    return fig, []

render_original_political_activity("partisan")
```

```python

def render_appendix_polarization(partisan_dimen):

    figsize = (10,9)
    fig = plt.figure(figsize=figsize)
    gs0 = gridspec.GridSpec(5, 6, figure=fig, 
                            width_ratios=[1,1,0.8,0.8,0.8,0.8],
                            height_ratios=[0.6,0.4,0.8,0.75,0.6])
    gs0.update(wspace=2, hspace=0)

    gs_a = gs0[0, 0:3]
    gs_b = gs0[0, 3:]
    
    gs_c = gs0[2, :2]
    gs_d = gs0[2, 2:]
    gs_e = gs0[4, 0:2]
    gs_f = gs0[4, 2:4]
    gs_g = gs0[4, 4:]
    #gs_e = gs0[4, :3]
    #gs_f = gs0[4, 3:]
    
    
    ax_dist = pol.render_partisan_distribution_difference_plot(fig, gs_a, partisan_dimen)
    plots.add_subplot_label(ax_dist, "a", x=-60)
    
    ax_prop = pol.render_proportion_extreme_activity_plot(fig, gs_b, partisan_dimen)
    ax_prop.margins(x=0)
    plots.add_subplot_label(ax_prop, "b", x=-60)
    
    ax_echo_null = pol.render_echo_chamber_plot(fig, gs_c, partisan_dimen, nullhyp=True)
    plots.add_subplot_label(ax_echo_null, "c", x=-60)
    
    ax_echo_community = pol.render_echo_chamber_community_plot(fig, gs_d, partisan_dimen)
    plots.add_subplot_label(ax_echo_community, "d", x=-60)
    
    ax_polprob = pol.render_polarization_heatmap(fig, gs_e, partisan_dimen, kind="correlation")
    plots.add_subplot_label(ax_polprob, "e", x=-60)
    
    
    ax_ccdf_1 = pol.render_polprob_ccdf(fig, gs_f, partisan_dimen, "sd_thresh")
    plots.add_subplot_label(ax_ccdf_1, "f", x=-45)
    
    ax_ccdf_2 = pol.render_polprob_ccdf(fig, gs_g, partisan_dimen, "mc_thresh")
    plots.add_subplot_label(ax_ccdf_2, "g", x=-60)
    
    #ax_confound_1 = pol.render_confound_plot(fig, gs_c, partisan_dimen, "month_since_join_index")
    #plots.add_subplot_label(ax_confound_1, "c", x=-50)
    
    #ax_confound_2 = pol.render_confound_plot(fig, gs_d, partisan_dimen, "active_month_index")
    #plots.add_subplot_label(ax_confound_2, "d", x=-60)
    
    return fig, []

render_appendix_polarization("partisan")
```
